```python
import requests
from bs4 import BeautifulSoup

l=[]
base_url = "https://www.flipkart.com/women/kurtas-kurtis/pr?sid=2oq%2Cc1r%2C3pj%2Cua6&otracker%5B%5D=categorytree&otracker%5B%5D=nmenu_sub_Women_0_Kurtas+and+Kurtis&page="
for i in range(1,11):
        b = base_url+str(i)
        print(b)
        r = requests.get(b)
    
        c = r.content
       
        soup= BeautifulSoup(c,"html.parser")

        all=soup.find_all("div",{"class":"_3O0U0u"})
       
        for a in all:
            d={}
            all1 = a.find_all("div",{"class":"_1vC4OE"})
            all2 = a.find_all("a",{"class":"_2cLu-l"})
            all3 = a.find_all("div",{"class":"VGWI6T"})
            all4 = a.find_all("span",class_='_3U7997')


            for i,j,k,m in zip(all1,all2,all3,all4):
                d['Price']=i.text
                d['Title']=j.text
                d['Price Off']=k.text
                d['Offer']=m.text

            l.append(d)

            
print(len(l))
import pandas
df = pandas.DataFrame(l)

df.to_csv('output_file.csv')
df
        



```
